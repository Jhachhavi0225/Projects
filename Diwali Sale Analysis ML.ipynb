```python
import matplotlib.pyplot as plt
```

```python
gender_counts=df['Gender'].value_counts() 
plt.figure(figsize=(6,6)) 
plt.pie(gender_counts,labels=gender_counts.index,autopct= '%1.1f%%',startangle = 140, colors=['skyblue','lightpink']) 
plt.title('Gender distribution') 
plt.show()
```

```python
marital_gender_counts =df.groupby(['Gender'	,	'Marital_Status']).size().unstack() 
print(marital_gender_counts)
```

```python

marital_gender_counts.plot(kind='bar', figsize=(8, 6), color=['skyblue', 'lightpink'])

# Add labels and title
plt.xlabel('Count of Gender')
plt.ylabel('Marital Status')
plt.title('Gender Distribution by Marital Status')

# Show the legend and plot
plt.legend(title='Gender')
plt.show()
```

```python
df['Age'].plot(kind="hist")
plt.title("Age Measure ") 
plt.xlabel("Age" ,fontsize=12)
plt.show()
```

```python
State_counts=df['State'].value_counts() 
plt.figure(figsize=(6,6)) 
plt.pie(State_counts,labels=State_counts.index,autopct= '%1.1f%%',startangle = 140, colors=['lightpink','purple']) 
plt.title('State ') 
plt.show() 

Zone_counts=df['Zone'].value_counts() 
plt.figure(figsize=(6,6)) 
plt.pie(Zone_counts,labels=Zone_counts.index,autopct= '%1.1f%%',startangle = 140, colors=['lightpink','purple']) 
plt.title('Zone popularity') 
plt.show()
```
